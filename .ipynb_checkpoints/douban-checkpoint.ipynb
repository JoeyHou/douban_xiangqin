### 1. 小组活跃度基本信息

```python
# def getGroupInfoDict(self, groupUrl, queryKeywordsList):
groupUrl = 'https://www.douban.com/group/345244/?ref=sidebar'

# 发送访问请求和接收
try:
    response = requests.get(groupUrl)
    response.encoding = 'utf-8'
except Exception as e:
    print(e)

# 读取响应内容并转换为soup对象
# html = response.read()
soup = bs4.BeautifulSoup(response.text, "lxml")


# 获取一个小组的所有信息
# 小组基本信息[9项]组来源、查询、组名、组ID、成员数、组地址、介绍、创建时间、组标签
# 小组活跃信息[3项]当天帖子总数、当天帖子历史累计回复数、平均帖子回复数
# 组长信息[3项]管理员姓名、全站唯一性ID、个人页面地址
# 系统信息[1项]表更新时间
groupInfoDict = {}

# 小组基本信息
groupInfoDict['groupSource'] = re.findall(r"www\.(.*)\.com", groupUrl)[0]
# groupInfoDict['groupQuery'] = ",".join(queryKeywordsList)
groupInfoDict['groupName'] = soup.title.string.strip()
groupInfoDict['groupId'] = re.findall(r'group/(.*)/', groupUrl)[0]

try:
    groupInfoDict['groupMemberNum'] = int(re.findall(r'浏览所有成员.*\([0-9]+\)', soup.text)[0].replace('浏览所有成员', '').replace('(', '').replace(')', ''))
except Exception as e:
    groupInfoDict['groupMemberNum'] = 0
#     logging.error(e)

# groupInfoDict['groupUrl'] = groupUrl
# groupInfoDict['groupIntro'] = str(soup.findAll("div", attrs={"class": "group-intro"})[0])
groupBoard = soup.find("div", attrs={"class": "group-board"}).p
groupInfoDict['groupCreateDate'] = re.findall(r"\d{4}-\d{2}-\d{2}", str(groupBoard))[0]

# 小组活跃信息
allPostInfoList = soup.find("table", attrs={"class":"olt"}).findAll("td", attrs={"class":"", "nowrap":"nowrap"})
currentDayCommentNumStrList = re.findall(r'<td class="" nowrap="nowrap">(\d*)</td>', str(allPostInfoList))
currentDayCommentNumIntList = map(lambda s: 0 if s == '' else int(s), currentDayCommentNumStrList)
groupInfoDict['currentDayCommentNum'] = sum(currentDayCommentNumIntList)
groupInfoDict['currentDayPostNum'] = len(currentDayCommentNumStrList)
groupInfoDict['currentDayAveCommentNum'] = int(float(groupInfoDict['currentDayCommentNum']/groupInfoDict['currentDayPostNum']))


# 表更新时间
groupInfoDict['tableUpdateDate'] = time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(time.time()))

pd.DataFrame([groupInfoDict])
```

```python


def find_last_comment(soup):
    '''Soup object -> Last comment time'''
    if soup.find('div', attrs = {'class': 'paginator'}):
        tmp_lst = list(soup.find('div', attrs = {'class': 'paginator'}).find_all('a'))
        new_url = tmp_lst[-2]['href']
        
        response = requests.get(new_url)
        response.encoding = 'utf-8'
        tmp_soup = bs4.BeautifulSoup(response.text, "lxml")
        
        return (tmp_soup.find_all('span', attrs = {'class': 'pubtime'})[-1].text)
    else:
        return (soup.find_all('span', attrs = {'class': 'pubtime'})[-1].text)
    
def find_post_like_n_comment_num(soup):
    '''Soup object -> number of like, number of comment'''
    tmp_str = soup.find('script', attrs = {'type': 'application/ld+json'})
    if tmp_str:
        tmp_str = tmp_str.text
        tmp_str = tmp_str.replace('\t', '').replace('\n', '').replace('\\', '')
        try:
            tmp_json = json.loads(str(tmp_str))
        except:
            return 0, 0
        like_num = tmp_json['interactionStatistic']['userInteractionCount']
        comment_num = tmp_json['commentCount']
        print('hi!')
        return int(like_num), int(comment_num)
    else:
        return 0, 0
    
def handle_single_post(soup, postUrl):
    '''Soup object, post url -> Group info dictionary'''
    try:
        response = requests.get(postUrl)
        response.encoding = 'utf-8'
    except Exception as e:
        print(e)

    # 读取响应内容并转换为soup对象
    soup = bs4.BeautifulSoup(response.text, "lxml")
    postContent = soup.find("div", attrs={"class":"topic-content clearfix"})
    postComment = soup.find("ul", attrs={"id":"comments", "class":"topic-reply"})

    # [3项]作者姓名、ID、签名、个人地址
    # [5项]内容、图片张数、图片地址列表字符串、作者评论、作者评论个数
    postDetailInfoDict = {}

    # Part.1 帖子基本信息 - 帖子链接、标题、ID、创建时间、最后回复时间、回复个数、喜欢人数
    postDetailInfoDict['postID'] = postUrl.split('/')[-2]
    postDetailInfoDict['postUrl'] = postUrl

    # 帖子标题
    try: # 查找是否存在长标题的标签并匹配
        postDetailInfoDict['postTitle'] = re.findall('<strong>标题：</strong>(.*)</td><td class="tablerc"></td></tr>', str(postContent))[0]
    except:
        postDetailInfoDict['postTitle'] = soup.title.text.strip()
    postDetailInfoDict['postCreateDate'] = str(postContent.find("span", attrs={"class":"color-green"}).string).strip()

    # 赞数 & 评论数
    like_num, comment_num = find_post_like_n_comment_num(soup)
    
    postDetailInfoDict["postLikeNum"] = like_num
    postDetailInfoDict['postCommentNum'] = comment_num

    # postLastCommentDate
    if postDetailInfoDict['postCommentNum'] > 0:
        postDetailInfoDict['postLastCommentDate'] = find_last_comment(soup)
    else:
        postDetailInfoDict['postLastCommentDate'] = postDetailInfoDict['postCreateDate']

    # Part.2
    # 作者姓名、ID、签名、个人地址
    postDetailInfoDict['postAuthorName'] = re.findall('alt="(.*)" class="pil"', str(postContent))[0]
    postDetailInfoDict['postAuthorUrl'] = re.findall('(https://www\.douban\.com/people/.*/)"><img', str(postContent))[0]
    postDetailInfoDict['postAuthorId'] = re.findall('https://www\.douban\.com/people/(.*)/"><img', str(postContent))[0]
        
    # Part.3 帖子内容
    # 帖子内容、所有评论、评论个数
    postDetailInfoDict['postContent'] = postContent.find("div", attrs={"class":"topic-content"}).text.replace("\r", "").replace("\n", "").replace(" ", "")

    return postDetailInfoDict
# postUrl = 'https://www.douban.com/group/topic/152524811/'
# handle_single_post(postUrl)
```

- 帖子回复（们）
    - 帖子ID
    - 回复内容
    - 回复者ID
    - 回复时间

```python
def handle_post_replies(soup, postUrl):
    all_reply = []
    postID = postUrl.split('/')[-2]
    current_url = postUrl
    while True:
        if current_url != postUrl:
            soup = get_soup(current_url)
        # Get all replies
        reply_lst = (soup.find_all('ul', attrs = {'class': 'topic-reply'})[-1].children)
        for reply in reply_lst:
            if reply == '\n':
                continue
            
            single_reply = {}

            # Post ID
            single_reply['postID'] = postID
            
            # Reply content
            single_reply['content'] = reply.find('p', attrs = {'class': 'reply-content'}).text
            
            # Poster id
            single_reply['posterID'] = reply.find('div', attrs = {'class': 'operation_div'})['id']
            
            # Reply time
            single_reply['replyTime'] = reply.find('span', attrs = {'class': 'pubtime'}).text
            
            # Add to all the replies
            all_reply.append(single_reply)

        # Check if it's the last page
        if soup.find('span', attrs = {'class': 'next'}) and soup.find('span', attrs = {'class': 'next'}).a:
            current_url = soup.find('span', attrs = {'class': 'next'}).a.get('href')
        else:
            break
    
    return pd.DataFrame(all_reply, columns = ['postID', 'posterID', 'replyTime', 'content'])
```

### 最终整合

```python
current_url = 'https://www.douban.com/group/345244/discussion?start=2550'
page = 103
start = time.time()
while True:
    soup = get_soup(current_url)
    print('===================')
    print('Getting page:', page)
    data_reply = []
    data_post = []
    
    # Get the replies and post info
    table = soup.find('table', attrs = {'class': 'olt'})
    post_num = 3
    if table:
        for post in table.children:
            if post == '\n':
                continue
            postUrl = post.find('a')
            if postUrl:
                postUrl = postUrl['href']
                tmp_soup = get_soup(postUrl)
                print(tmp_soup.find_all('span', attrs = {'class': 'pubtime'})[-1].text)
                print('Getting the post:', post_num)
                data_reply.append(handle_post_replies(tmp_soup, postUrl))
                data_post.append(handle_single_post(tmp_soup, postUrl))
                post_num += 1
                break
#                 if post_num == 5:
#                     break

    # Check if it's the last page
    if soup.find('span', attrs = {'class': 'next'}) and soup.find('span', attrs = {'class': 'next'}).a:
        pd.concat(data_reply).to_csv('reply_' + str(page) + '.csv', index = False)
        pd.DataFrame(data_post).to_csv('post_' + str(page) + '.csv', index = False)
        current_url = soup.find('span', attrs = {'class': 'next'}).a['href']
        page += 1
    else:
        print('========Done!========')
        break
    
    break
end = time.time()
print(end - start)
```

```python
def get_new_ip():
    print('Getting new IP!')
    order = "d9688431f712b21b42c2f103dca284d4"
    apiUrl = "http://api.ip.data5u.com/dynamic/get.html?order=" + order;
    return {'https': requests.get(apiUrl).content.decode().strip()}

access_count = 0
proxy = get_new_ip()
def get_soup(url, user_agent):
    '''URL -> return soup object'''
    # print('Getting soup..')
    global proxy
    global access_count
    access_count += 1
    if access_count > 35:
        proxy = get_new_ip()
        access_count = 0
 
    headers = {
        'User-Agent': user_agent,
    }

    while True:
        try:
            response = requests.get(url, headers = headers, proxies = proxy, timeout = (5, 20))
            response.encoding = 'utf-8'
            soup = bs4.BeautifulSoup(response.text, "lxml")
            if soup.find('p') and 'IP' in soup.find('p').text:
                continue
            else:
                break
        except Exception as e:
            print(e)
            proxy = get_new_ip()
            continue
    return soup

def get_author_info(soup):
    try:
        gender = (soup.find('span', attrs = {'class': 'gender'})['class'])[1]
    except:
        gender = ''
    
    try:
        location = soup.find('div', attrs = {'class': 'loc'}).text
    except:
        location = ''
    
    return gender, location
```

### 词频分析

```python
import os, codecs
import jieba
from collections import Counter
import sys
data_path = './data/post_and_reply/'
file_lst = os.listdir(data_path)
```

```python
def get_words(txt):
    seg_list = jieba.cut(txt)
    c = Counter()
    for x in seg_list:
        if len(x) > 1 and x != '\r\n':
            c[x] += 1
    result = []
    for (k,v) in c.most_common(100):
        result.append({'Word': k, 'Count': v})
    return pd.DataFrame(result, columns = ['Word', 'Count'])

def concat_all_reply():
    all_reply = []
    for file in file_lst:
        if 'reply' in file and '.csv' in file:
            try: 
                tmp_df = pd.read_csv(data_path + file, lineterminator='\n')
            except:
                print(file)
            try:
                all_reply.append(get_words(' '.join(tmp_df.dropna().content.values)))
            except:
                display(tmp_df)
    return pd.concat(all_reply)

def process_content(s):
    if s.postContent.find(s.postCreateDate.replace(' ', '')) >= 0 or s.postContent[:2] == '来自':
        start_idx = s.postContent.find(s.postCreateDate.replace(' ', '')) + 18
    else:
#         print(s.postContent)
        start_idx = 0
    s.postContent = s.postContent[start_idx:]
    return s
def concat_all_post():
    all_post = []
    all_title = []
    for file in file_lst:
        if 'post' in file and '.csv' in file:
            try: 
                tmp_df = pd.read_csv(data_path + file)
            except:
                print(file)
            try:
                new_df = tmp_df.dropna().apply(process_content, axis = 1)
                all_post.append(get_words(' '.join(new_df.dropna().postContent.values)))
                all_title.append(get_words(' '.join(tmp_df.dropna().postTitle.values)))
            except:
                display(tmp_df)
    return pd.concat(all_post), pd.concat(all_title)
```

```python
post, title = concat_all_post()
reply = concat_all_reply()
```

```python
post = post.reset_index(drop = True).groupby('Word').agg(sum).sort_values(by = 'Count', ascending = False)
post.to_csv('post_summary.csv')
post = post.rename(columns = {'Count': 'Post'})
```

```python
title = title.reset_index(drop = True).groupby('Word').agg(sum).sort_values(by = 'Count', ascending = False)
title.to_csv('title_summary.csv')
title = title.rename(columns = {'Count': 'Title'})
```

```python
reply = reply.reset_index(drop = True).groupby('Word').agg(sum).sort_values(by = 'Count', ascending = False)
reply.to_csv('reply_summary.csv')
reply = reply.rename(columns = {'Count': 'Reply'})
```

```python
all_count = reply.join(post, how = 'outer').join(title, how = 'outer').fillna(0)
```

```python
all_count['Total'] = all_count.apply(lambda s: s.Reply + s.Post + s.Title, axis = 1)
all_count = all_count.sort_values(by = 'Total', ascending = False)
all_count.to_csv('all_word_count.csv')
```

### 时间问题
- 每天发帖数量
- 每天回复数量

```python
reply_lst = []
post_lst = []
for file in file_lst:
    if 'reply' in file:
        reply_lst.append(pd.read_csv('./data/' + file, lineterminator='\n')[['replyTime']])
    if 'post' in file:
        post_lst.append(pd.read_csv('./data/' + file)[['postCreateDate']])
```

```python
all_reply_time = pd.concat(reply_lst).reset_index(drop = True)
all_post_time = pd.concat(post_lst).reset_index(drop = True)
```

```python
def strip_time(t):
    return t[:-9]
all_reply_time.replyTime = all_reply_time.replyTime.apply(strip_time)
all_reply_time['count'] = 1
all_post_time.postCreateDate = all_post_time.postCreateDate.apply(strip_time)
all_post_time['count'] = 1
```

```python
all_reply_time.groupby('replyTime').agg(sum).to_csv('reply_time.csv')
```

```python
all_post_time.groupby('postCreateDate').agg(sum).to_csv('post_time.csv')
```
